```python
import os
import joblib
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
```

```python
from google.colab import drive
drive.mount('/content/drive')

data_prep_path_final='/content/drive/MyDrive/dataPrepFinal/code'
```

```python
X_train = joblib.load(f"{data_prep_path_final}/x_train.pkl")
X_test = joblib.load(f"{data_prep_path_final}/x_test.pkl")
y_train = joblib.load(f"{data_prep_path_final}/y_train.pkl")
y_test = joblib.load(f"{data_prep_path_final}/y_test.pkl")
y_classes = joblib.load(f"{data_prep_path_final}/y_classes.pkl")
```

```python
from sklearn.multiclass import OneVsRestClassifier
```

```python
mb_classifier = MultinomialNB()

clf = OneVsRestClassifier(mb_classifier)
clf.fit(X_train, y_train)
```

```python
joblib.dump(clf, f"{data_prep_path_final}/mb_classifier.pkl")
```

```python
y_pred = clf.predict(X_test)
```

```python
clf_baseline_mb = joblib.load(f"{data_prep_path_final}/mb_classifier.pkl")
```

```python
from sklearn.metrics import precision_recall_fscore_support as score

precision, recall, fscore, support = score(y_test, y_pred)

print(f"precision: {precision}")
print(f"recall: {recall}")
print(f"fscore: {fscore}")
print(f"support: {support}")
```

```python
from sklearn.metrics import hamming_loss

hamming = []

for i, (test, pred) in enumerate(zip(y_test.T, y_pred.T)):
    hamming.append(hamming_loss(test, pred))
```

```python
metric_df = pd.DataFrame(data=[precision, recall, fscore, hamming, support],
                         index=["Precision", "Recall", "F-1 score", "Hamming loss", "True count"],
                         columns=y_classes)
```
